```python
#Normalizing the data
df = (df - df.mean()) / df.std()
print()
print('Normalized Data')
print(df.head(10))
```

```python
#Features
X=df.copy()
X = X.drop(columns=['Date', 'Rented Bike Count','Dew point temperature(°C)'])
X.head()
```

```python
#Target
y=df['Rented Bike Count']
y.head()
```

```python
#Apply Train-Test-Spilt in our dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, shuffle=False, random_state=42)
```

# Model Implementation

```python
#Import Libraries
import torch
import torch.nn as nn
```

```python
X_train = torch.tensor(np.array(X_train).astype(np.float32))
X_test = torch.tensor(np.array(X_test).astype(np.float32))
y_train = torch.tensor(np.array(y_train).astype(np.float32))
y_test = torch.tensor(np.array(y_test).astype(np.float32))
```

```python
#Implement NN model called dfNN
class dfNN(nn.Module):
    def __init__(self):
        super(dfNN, self).__init__()
        self.fc1 = nn.Linear(11,63)
        self.fc2 = nn.Linear(63,1)
        self.sigmoid = torch.nn.Sigmoid()
        
    def forward(self, x):
        hidden = self.fc1(x)
        hidden = self.sigmoid(hidden)
        hidden = self.fc2(hidden)
        output = self.sigmoid(hidden)
        return output
```

```python
df_model = dfNN()
```

```python
#Specify loss function and optimising it using Stochastic Gradient Descent(SGD)
criterion = torch.nn.MSELoss()
optimizer = torch.optim.SGD(df_model.parameters(), lr = 0.001)
```

```python
#Checking test loss before training
print(df_model.eval())
y_pred = df_model(X_test)
before_train = criterion(y_pred.squeeze(), y_test)
print('Test loss before training' , before_train.item())
```

```python
#Implementation

import time
t0 = time.time()
acclist = [] #List to get accuracy at each epoch run 
train_losses= [] #List to get train loss at each epoch run 
time1 = [] #List to get training time at each epoch run 
epoch = 100
for epoch in range(epoch):
    correct = 0
    optimizer.zero_grad()
    y_pred = df_model(X_train)
    
    loss = criterion(y_pred.squeeze(), y_train)
    train_losses.append(loss.item())
    print('Epoch {}: train loss: {}'.format(epoch, loss.item()))
    accuracy = (torch.round(y_pred, decimals = 0) == torch.round(y_test, decimals = 0)).float().mean()*100
    print(f'Test accuarcy: {accuracy:.2f}')
    acclist.append(accuracy)
    time1.append((time.time()-t0))
    loss.backward()

    optimizer.step()
    
print("Training time:", time.time()-t0)    
```

```python
#Visualizing train time vs epoch
plt.plot(time1)
plt.xlabel('Epoch')
plt.ylabel('Time taken in seconds')
plt.title('LR 0.001, Nodes 63, Epoch 100')
plt.show()
```

```python
#Visualizing train loss vs epoch
plt.plot(train_losses)
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title('LR 0.001, Nodes 63, Epoch 100')
plt.show()
```

```python
#Visualizing accuracy vs epoch
plt.plot(acclist)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('LR 0.001, Nodes 63, Epoch 100')
plt.show()
```
